### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
# x = ["A", "B", "C", "D"]
# y = [1, 2, 3, 4]
sub = ["s1","s2","s3","s4"]
Marks = [20,10,67,90]
plt.bar(sub, Marks)
 
for index, value in enumerate(sub):
    plt.text(value, index,
             str(value))
 
plt.show()
```
